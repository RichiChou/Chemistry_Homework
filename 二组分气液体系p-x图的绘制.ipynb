```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
xb_l = np.array([0, 0.1, 0.2, 0.4, 0.6, 0.8, 0.95, 0.98, 1]) # 液相B组分摩尔比
pa = np.array([2.90, 2.59, 2.37, 2.07, 1.89, 1.81, 1.44, 0.67, 0]) # A蒸汽压
pb = np.array([0, 1.08, 1.79, 2.65, 2.89, 2.91, 3.09, 3.13, 3.17]) # B蒸汽压
```

```python
p_total = pa + pb
p_total
```

```python
# 绘制蒸汽分压即总压p-x图
plt.plot(xb_l, pa, label="Pa")
plt.plot(xb_l, pb, label="Pb")
plt.plot(xb_l, p_total, label="P_total")
plt.legend()
plt.show()
```

```python
# 计算气相中B组分的摩尔比
xb_g = pb / (pa + pb)
xb_g
```

```python
plt.plot(xb_l, p_total, label="Liquid Line")
plt.plot(xb_g, p_total, label="Gaseous Line")
plt.legend()
plt.show()
```
